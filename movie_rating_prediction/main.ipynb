```python
import pandas as pd
```

```python
data = pd.read_csv('IMDb Movies India.csv',encoding='ISO-8859-1')
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
data['Year'].dtype
```

```python
data['Year'] = data['Year'].str.extract(r'(\d+)').astype(int)
```

```python
data = data.drop("Name",axis=1)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Encoding genre, director, and actors
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(data[['Genre', 'Director', 'Actor 1','Actor 2','Actor 3']])
```

```python
encoded_df = pd.DataFrame(encoded_features.toarray(), columns=encoder.get_feature_names_out(['Genre', 'Director', 'Actor 1','Actor 2','Actor 3']))

# Combine the encoded features with the original dataset
data_encoded = pd.concat([data, encoded_df], axis=1).drop(['Genre', 'Director', 'Actor 1','Actor 2','Actor 3'], axis=1)

# Display the first few rows of the processed dataset
data_encoded.head()
```
